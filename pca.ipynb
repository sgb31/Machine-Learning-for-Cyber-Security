```python
import pandas as pd
```

```python
df =pd.read_csv("Cleaneddata.csv")
```

```python
data = df.drop(["Timestamp","Unnamed: 0","IncidentGrade"],axis=1)
```

```python
data.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
scaleddata = sc.fit_transform(data) 
```

```python
scaleddata = pd.DataFrame(scaleddata,columns = data.columns)
```

```python
scaleddata.head()
```

```python
pca = PCA(n_components=15)
pcaeddata = pca.fit_transform(scaleddata)
```

```python
pcaeddf = pd.DataFrame(pcaeddata)
```

```python
pcaeddf
```

```python
y = df['IncidentGrade']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(pcaeddf,y,test_size=0.3,random_state=42)
```

```python
lr = LogisticRegression(multi_class = 'multinomial')
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```
